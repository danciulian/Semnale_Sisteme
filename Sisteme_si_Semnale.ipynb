# Semnale și sisteme

## 1. Procesarea semnalelor. Clasificarea semnalelor
Un *semnal* este o funcție care reprezintă o cantitate fizică care conține informații despre comportamentul sau natura unui fenomen. Din punct de vedere matematic, un semnal este reprezentat de o funcție având ca argument în mod uzual timpul $t$. 
Procesarea semnalelor poate fi studiată din punct de vedere analog(continuu) sau din punct de vedere digital(discret).

**Semnalul continuu** este un semnal care variază în mod continuu și poate lua orice valoare într-un interval specificat. Caracteristica principală a unui astfel de semnal este continuitatea în timp, mai exact, semanlul continuu este definit pentru orice moment de timp din intervalul său de existență - nu există întreruperi sau *salturi* în valorile sale în timp.

**Semnalul discret** este un tip de semnal care ia valori la anumite intervale de timp. În cele ce urmează ne vom concentra pe studiul semnalelor discrete, apelând la nevoie de tehnicile utilizate în spațiul continuu. 

### Eșantionare
Semnalele continue sunt transformate adesea în semnale discrete printr-un proces numit eșantionare. Acest proces implică măsurarea valorilor semnalului la intervale regulate de timp. Frecvența de eșantionare determină cât de des sunt preluate aceste măsurători și înfluențează calitatea semnalului discret rezultat.

Avantajul folosirii semnalelor discrete este disponibilitatea algoritmilor și tehnicilor digitate de procesare și analiză. Astfel, acest lucru permite implementarea de *filtre*, *transformate* și alte operații complexe care sunt esențiale în domenii precum telecomunicațiile digitale. 

Din punct de vedere matematic, diferentele dintre discret și continuu sunt:
1. Calculul integral este înlocuit de sume
2. Calculul diferențial este înlocuit de diferențe finite
3. Funcția Dirac este înlocuită de impulsul 

Eșantionarea unui semnal continuu ${x(t)}$ la intervale regulate de timp duce la creearea unui semnal discret ${x[n]}$ sau matematic scris: $$\displaystyle x[n]=x(nT_{s}), \text{ } (\forall) n \in \mathbb{Z}, \text{ unde}$$ 
1. $T_{s}$ este perioada de eșantionare
2. $\displaystyle f_{s}=\frac{1}{T_{s}}$, este **frecvența** de eșantionare.

**Frecveța** este măsura numărului de repetări ale unui fenomen periodic în unitatea de timp. Unitatea de măsură pentru frecvență este hertz. Pentru un anumit fenomen fizic care se repetă la un interval de timp $T$, se poate calcula frecvența lui folosind formula mai sus amintită.

```python
%matplotlib widget
from ipywidgets import interact
import ipywidgets as widgets

import matplotlib.pyplot as plt
import numpy as np

def plot(freq):
    x = np.linspace(0, 2*np.pi)
    y = np.sin(x * freq)
    plt.plot(x, y)


interact(plot, freq = widgets.FloatSlider(value=2, min=0.1, max=5, step=0.1))
```

## 2. Semnale de bază în timp discret

### 2.1 Șirul pas unitate

Șirul pas unitate este definit după cum urmează:

$$u[n] = 
\begin{cases}
1, \text{ } n \geq 0\\
0, \text{ } n < 0
\end{cases}
$$

În mod similar se poate defini șirul shiftat în timp:

$$u[n-k] = 
\begin{cases}
1, \text{ } n \geq k\\
0, \text{ } n < k
\end{cases}
$$

### 2.2 Șirul impulsul unitate

$$\delta[n] = 
\begin{cases}
0, \text{ } n \neq 0\\
1, \text{ } n=0
\end{cases}
$$

în mod asemănător definim șirul shiftat în timp:
$$\delta[n-k] = 
\begin{cases}
0, \text{ } n \neq k\\
1, \text{ } n=k
\end{cases}
$$

Se poate observa că următoarele relații sunt adevărate:

* $$x[n]\delta[n-k] = x[k]\delta[n-k]$$
* $$\delta[n] = u[n] - u[n-1]$$
* $$u[n] = \sum^{n}_{k = -\infty}{\delta[k]} = \sum^{\infty}_{k = 0}{\delta[n-k]}$$
* $$x[n] = \sum^{\infty}_{k = -\infty}{x[k]\delta[n-k]}.$$

### 2.3 Șirul complex exponențial
Considerăm șirul $x[n]=z^{n}$, $(\forall) n \in \mathbb{N}$ și $z \in \mathbb{C}$. Folosind formula lui Euler $e^{i\Omega}=\cos{\Omega}+\sin{\Omega}$, putem rescrie $$x[n]=\left|z\right|^{n}\left(\cos(\Omega n)+i\sin(\Omega n)\right),$$ unde $\Omega$ este frecvența. 

### 2.4 Șirul general complex exponențial

Acest șir este definit ca $$x[n]=C\alpha^{n},$$ unde $C, \alpha \in \mathbb{C}$.

### 2.5 Șirul real exponențial

Dacă $C$ și $\alpha$ din relația de mai sus sunt amândouă reale, atunci șirul $x[n]$ este un șir real exponențial. Se pot distinge patru cazuri: $\alpha >0$, $0 < \alpha < 1$, $-1<\alpha<0$ și $\alpha < -1$. De remarcat faptul că dacă $\alpha = 0$, atunci șirul $x[n]$ este un șir constant, ori dacă $\alpha = -1$ atunci șirul va lua valorile $C$ sau $-C$.


### 2.6 Șirul sinusoidal

Este exprimat ca fiind $$x[n] = A\cos(\Omega n + \theta)$$

Sinusoidele în timp discret sunt de forma $\cos((\Omega + 2\pi k)n)=\cos(\Omega n)$, pentru orice $k \in \mathbb{Z}$.
   
   a) Frecvențele în timp discret $\Omega$ și $\Omega+2\pi k$ generează șiruri identice pentru orice $k \in \mathbb{Z}$.

   b) Un șir este $x$ este denumit periodic dacă există $N>0$ cu proprietatea că $x[n+N]=x[n]$.

   c) Din relația de mai sus se poate deduce că $x[n+mN]=x[n]$, pentru orice $m\in \mathbb{Z}$.

   d) cel mai mic număr $N$ pentru care $x[n+N]=x[n]$ se numește perioadă fundamentală.


## 3. Sisteme și clasificarea sistemelor

### Reprezentarea sistemelor

Un sistem este un model matematica care caracterizeaza un proces fizic care transformă un semnal de tip *input* în unul de tip *output*. Fie $x$ semnalul input și $y$ semnalul output. Atunci sistemul poate fi înțeles ca fiind transformarea lui $x$ în $y$ prin intermediul unui operator $T$. Scriind sub o formă mai ușoară putem rezuma că: $$y = Tx.$$

În cele ce urmează vom expune câteva tipuri de sisteme, precum și câteva dintre caracteristicile lor. 

### 3.1 Sisteme deterministice și stochastice
Un semnal se numește determinat, dacă valorile sau sunt cunoscute în orice moment de timp. Dacă semnalul input și semnalul output sunt semnale bine determinate, atunci sistemul se numește sistem **determinat**.

Dacă semnalul input și semnalul output sunt semnale aleatoare atunci sistemul poartă numele de sistem **stochastic**.
### 3.2 Sisteme continue și discrete

Dacă semnalul input și output sunt semnale continuue, atunci sistemul se numește **sistem  continuu în timp**.
În mod analog, tragem concluzia că dacă semnalul input și output sunt discrete, atunci sistemul se numește **sistem discret în timp**. 
### 3.3 Sisteme cu și fără memorie
Un sistem se numește sistem fără memorie dacă semnalul output la un anumit moment de tip depinde doar de semnalul input la același moment de timp. 

Un exemplu de sistem cu memorie este sistemul discret ale cărui semnale de input și output respectă relația: $$y[n]=\sum_{k = -\infty}^{n}{x[k]}.$$

### 3.4 Sisteme cauzale și non-cauzale

Un sistem se numește cauzal dacă semnalul output la momentul prezent depinde doare de valorile prezente sau trecute ale semnalului input. 

Un sistem se numește non-cauzal (sau anticipativ) dacă semnalul output la momentul prezent depinde the valorile viitoare ale semnalului input. Următoarul sistem este un exemplu de sistem non-cauzal: $$y[n]=x[-n].$$

##### Remarcă
Toate sistemele fără memorie sunt sisteme cauzale. Reciproca este falsă.
### 3.5 Sisteme liniare și neliniare
Dacă un operator $T$ satisface următoarele două condiții 
**<dd>1. Adititvitate</dd>**
Fiind date $Tx_{1}=y_{1}$ și $Tx_{2}=y_{2}$, atunci $\displaystyle T\{x_1 + x_2\}=y_1+y_2$ pentru orice semnale input $x_1$ și $x_2$.

**<dd>2. Scalare</dd>**

$\displaystyle T\{\alpha x\}=\alpha y, $ pentru orice semnal input $x$ și orice scalar $\alpha$.


atunci $T$ se numește operator liniar, iar sistemul reprezentat de operatorul $T$ se numește *sistem liniar*.

Cele două condiții de mai sus pot fi rescrise sub forma:

$$T\{\alpha_{1}x_1+\alpha_{2}x_2\}=\alpha_{1}y_1+\alpha_{2}y_{2}, $$ unde $\alpha_{1}$, $\alpha_{2}$ sunt numere scalare arbitrare. Această relație poartă denumirea de **proprietatea de superpoziție**.

### 3.6 Sisteme dependente și independente de timp

Dacă semnalul input și semnalul output nu se modifică odată cu timpul, atunci semnalul este denumit **semnal independent de timp**. Matematic, putem rescrie: $$T\{x[n-k]\}=y[n-k].$$

### 3.7 Sisteme stabile

Un sistem este stabil dacă din $\left|x\right|<k_1$ rezultă că $\left|y\right|<k_2$, unde $x$ este semnalul input, $y$ semnalul output, iar $k_1, k_2$ sunt două constante finite reale. 

Sistemul $y[n]=(n+1)u[n]$, unde semnalul input $x[n]=u[n]$, iar $u[n]$ este șirul pasul unitate, este sistem instabil, deoarece semnalul output $y[n]$ crește odată cu creșterea lui $n$.
### 3.8 Sisteme de tip feedback

În sistemele de tip feedback, semnalul output este refolosit și adăguat la semnalul input. 

### 3.9 Sisteme LTI (liniare invariante în timp)

Dacă sistemul este liniar și de asemenea independent de timp, atunci el se numește sistem liniar independent de timp. În continunare îl vom denumi **sistem LTI**.

## 4. Sisteme liniare invariante în timp (Sisteme LTI)

### 4.1 Răspunsul unitate
Răspunsul unitate $h[n]$ a unui sistem discret LTI (reprezentat de operatorul $T$) este definit ca fiind $$h[n]=T\{\delta[n]\},$$ unde semnalul input este $\delta[n]$.

### 4.2 Răspunsul la un semnal input arbitrar
Pe de altă parte, orice semnal se poate scrie:

$$x[n]=\sum_{k=-\infty}^{+\infty}{x[k]\delta[n-k]}$$

Vom cosidera semnalul $y[n]$ ca fiind răspunsul sistemului $T$ asupra semnalului $x[n]$.

\begin{eqnarray}
y[n] &=& T\left\{x[n]\right\}\\
&=& T\left\{\sum^{\infty}_{k=-\infty}{x[k]\delta[n-k]}\right\}\\
&=& \sum^{\infty}_{k=-\infty}{x[k]T\left\{\delta[n-k]\right\}}\\
&=& \sum^{\infty}_{k=-\infty}{x[k]h[n-k]}.
\end{eqnarray}
Această ultimă ecuație pune în evidență faptul că orice sistem discret LTI este caracterizat de răspunsul impuls $h[n]$.
#### 4.2.1 Convoluție

Ultima parte a ecuației anterioare poartă numele de convoluție și se notează $x*h$
Astfel, putem rescrie:

$$y[n]=x[n]*h[n]=\sum^{\infty}_{k=-\infty}{x[k]h[n-k]}.$$

Putem concluziona că semnalul output al oricărui sistem discret LTI este convoluția dintre semnalul input $x[n]$ și răspunsul unitate $h[n]$ al sistemului.

#### 4.2.2 Proprietățile convoluției
1. Comutativitate
   $$x[n]*h[n]=h[n]*x[n]$$
2. Asociativitate
   $$\left\{x[n]*h_1[n]\right\}*h_2[n] = x[n]*\left\{h_{1}[n]*h_{2}[n]\right\}$$
3. Distributivitate
   $$x[n]*\left\{ h_{1}[n]+h_{2}[n]\right\}=x[n]*h_{1}[n]+x[n]*h_{2}[n]$$

### 4.3 Răspunsul frecvență
În continuare considerăm un sistem discret in timp de tip LTI cu răspunsul unitate $h[n]$ și un semnal input $x[n]=e^{i\omega n}$. Avem, deci:
\begin{eqnarray}
y[n] &=& \sum^{\infty}_{k=-\infty}{h[k]x[n-k]}\\
&=& \sum^{\infty}_{k=-\infty}{h[k]e^{i\omega(n-k)}}\\
&=& \sum^{\infty}_{k=-\infty}{h[k]e^{-i\omega k}e^{i\omega n}}\\
&=& e^{i\omega n}\sum^{\infty}_{k=-\infty}{h[k]e^{-i\omega k}}\\
&=&e^{i\omega n}H\left(e^{i\omega}\right),
\end{eqnarray}
unde $H\left(e^{i\omega}\right)$ este un număr complex care reprezintă **răspunsul frecvență** a sistemul de tip LTI la o frecvență $\omega$.

Cum $H\left(e^{i\omega}\right)$ este un număr complex, îl putem rescrie ca sumă dintre partea reală și partea imaginară:
$$H\left(e^{i\omega}\right)=H_{R}\left(e^{i\omega}\right)+i\cdot H_{I}\left(e^{i\omega}\right)$$ sau în termeni de *magnitudine* și *fază* ca $$H\left(e^{i\omega}\right)=\left|H\left(e^{i\omega}\right)\right|e^{i\angle H\left(e^{i\omega}\right)}$$
Revenind, $y[n]$ se rescrie sub următoarea formă:
\begin{eqnarray}
y[n]&=& e^{i\omega n}\left|H\left(e^{i\omega}\right)\right|e^{i\angle H\left(e^{i\omega}\right)}\\
&=&\left|H\left(e^{i\omega}\right)\right|e^{i\left(\omega n + \angle H\left(e^{i\omega}\right)\right)}.
\end{eqnarray}

Deducem că output-ul sistemului este tot de tip exponențial complex, dar cu valori diferite ale magnitudinii și fazei. 
În continuare considerăm șirul $x[n]=A\cos{(\omega n + \phi)}, \text{ } (\forall) n \in \mathbb{Z}$. Folosind identitatea lui Euler, putem rescrie: 
\begin{eqnarray}
A\cos{(\omega n + \phi)} &=& \frac{A}{2}\left(e^{j(\omega n + \phi)}+e^{-j(\omega n + \phi)}\right)\\
&=&\frac{A}{2}e^{j\phi}e^{j\omega n} + \frac{A}{2}e^{-j\phi}e^{-j\omega n}. 
\end{eqnarray}

Trecând un semnal $x[n]$ printr-un sistem $y[n]$ obținem: 
\begin{eqnarray}
y[n] &=& \frac{A}{2}e^{j\phi}H\left(e^{j\omega}\right)e^{j\omega n} + \frac{A}{2}e^{-j\phi}H\left(e^{-j\omega}\right)e^{-j\omega n}\\
&=& \frac{A}{2}e^{j\phi+\angle H\left(e^{j\omega}\right)}\left|H\left(e^{j\omega}\right)\right|e^{j\omega n} + \frac{A}{2}e^{-j\phi+\angle H\left(e^{-j\omega}\right)}\left|H\left(e^{-j\omega}\right)\right|e^{-j\omega n}.
\end{eqnarray}
Având în vedere că $\displaystyle \left|H\left(e^{-j\omega}\right)\right|=\left|H\left(e^{j\omega}\right)\right|$ și că $\displaystyle \angle H\left(e^{-j\omega}\right)=-\angle H\left(e^{j\omega}\right)$, relația de mai sus se poate rescrie după cum urmează:
$$y[n] = \left|H\left(e^{j\omega}\right)\right|A\cos{\left(\omega n + \phi + \angle H\left(e^{j\omega}\right)\right)}.$$
Folosind notația $\theta(\omega)=H\left(e^{j\omega}\right)$ atunci putem rescrie: 
\begin{eqnarray}
y[n] &=& \left|H\left(e^{j\omega}\right)\right|A\cos\left(\omega\left(n+\theta(\omega)/\omega\right)+\phi\right)
&=&\left|H\left(e^{j\omega}\right)\right|A\cos\left(\omega\left(n-\tau_{p}\left(\omega\right)\right)+\phi\right),
\end{eqnarray}
unde $\displaystyle \tau_{p}=-\frac{\theta\left(\omega\right)}{\omega}$ poartă numele de *faza de întârziere*.

Se observă că am pornit de la un semnal input $x[n]=A\cos(\omega n +\phi)$, $(\forall) n \in \mathbb{Z}$ și am ajuns la un semnal output cu aceeași sinusoidă, dar cu două diferențe majore: 

* Amplitudinea scalată: $\left|H\left(e^{j\omega}\right)\right|$
* Faza shiftată: $\angle H\left(e^{j\omega}\right)$


## 5. Sisteme descrise de ecuații cu diferență(LCCDE)

Ecuațiile cu diferență (în engleză *Linear Constant Coefficient Difference Equations*, prescurtat **LCCDE**) sunt omoloagele ecuațiilor diferențiale dintr-un domeniu continuu. O ecuație liniară cu diferențe și coefiecienți constanți are forma următoare:

$$\sum^{N}_{k=0}{a_{k}y[n-k]}=\sum^{M}_{k=0}{b_{k}u[n-k]}, \text{ } a_{k}, b_{k} \in \mathbb{R} \text{ și } M,N \in \mathbb{Z}_{+}.$$

Presupunând că sistemul este cauzal și că $a_{0} \neq 0$, ecuația de mai sus se rescrie:
$$y[n]=\frac{1}{a_{0}}\left(\sum^{M}_{k=0}{b_{k}u[n-k]} - \sum^{N}_{k=1}{a_{k}y[n-k]}\right).$$

## 6. Transformata $z$ și sistemele LTI discrete

Transformata **$z$** permite studierea semnalelor discrete să fie analizate și din punct de vedere al frecvenței.

Fiind dat un șir $x[n]$, transformata sa **$z$** este definită ca fiind:
$$\mathcal{Z}\{x[n]\}=X(z)=\sum^{\infty}_{n=-\infty}{x[n]z^{-n}}, \text{ }z \in \mathbb{C}.$$

Considerăm șirul $x[n]$ definit după cum urmează:
$$x[n]=
\begin{cases}
a^{n}, \text{ }n \geq 0, a \in \mathbb{R}, a \neq 0\\
0, \text{ în caz contrar}
\end{cases}
$$
și transformata $z$:
$$X(z)=\sum^{\infty}_{n=0}{a^{n}z^{-n}}=\sum^{\infty}_{n=0}{\left(\frac{a}{z}\right)^{n}}.$$
Se poate demonstra că suma anterioară este convergentă dacă și numai dacă $\displaystyle \left|\frac{a}{z}\right| < 1$. În acest caz putem concluziona $$X(z) = \frac{z}{z-a}.$$

Acum, considerăm șirul:
$$x[n]=
\begin{cases}
-a^{n}, \text{ }n < 0, a \in \mathbb{R}, a \neq 0\\
0, \text{ în caz contrar}
\end{cases}
$$
și transformata $z$:
$$X(z)=-\sum^{-1}_{n=-\infty}{a^{n}z^{-n}}=-\sum^{\infty}_{n=1}{a^{-n}z^{n}}=-\sum^{\infty}_{n=1}{\left(\frac{z}{a}\right)^{n}}.$$
În mod analog se demonstrează că suma converge dacă și numai dacă $\displaystyle \left|\frac{z}{a}\right| < 1$. La fel, concluzionăm că $$X(z) = \frac{z}{z-a}.$$

## 7. Funcția de transfer a sistemelor LTI discrete

Pentru un sistem LTI cu răspunsul sistemului $h[n]$ asupra semnalului impulsul unitate, avem următoarea relație:
$$y[n]=u[n] \star h[n] \leftrightarrow Y(z)=U(z)H(z)$$
sau
$$H(z)=\frac{Y(z)}{U(z)}.$$

$H(z)$ este denumită funcția de transfer a sistemului.

## 8. Funcția de transfer a sistemlor LTI caracterizate de LCCDE(diferențele de ecuații)

Considerăm un sistem descris de LCCDE: $$\sum^{N}_{k=0}{a_{k}y[n-k]}=\sum^{M}_{k=0}{b_{k}u[n-k]}.$$ Folosind în continuare transformata $z$, putem rescrie: 

$$\sum^{N}_{k=0}{a_{k}y[n-k]}=\sum^{M}_{k=0}{b_{k}u[n-k]} \leftrightarrow \sum^{N}_{k=0}{a_{k}z^{-k}Y(z)}=\sum^{M}_{k=0}{b_{k}z^{-k}U(z)},$$ relație echivalentă cu $$\frac{Y(z)}{U(z)}=\frac{b_{0}+b_{1}z^{-1}+\ldots + b_{M}z^{-M}}{a_{0}+a_{1}z^{-1}+\ldots + a_{N}z^{-N}}=H(z).$$

Expresia de mai sus este o expresie polinomială și poate fi rescrisă sub următoarea formă: 

\begin{eqnarray}
H(s) &=& k\frac{(s-z_{1})(s-z_{2})\cdot \dots \cdot (s-z_{m})}{(s-p_{1})(s-p_{2})\cdot \dots \cdot (s-p_{n})}.
\end{eqnarray}

S-a făcut schimbarea de variabilă $z \longrightarrow s$ pentru a pune mai bine în evidență următoarele lucruri:
* elementele mulțimii $\displaystyle \{z_{1}, z_{2}, \ldots, z_{m}\}$ poartă numele de *zerourile* polinomului $H$.
* elementele mulțimii $\displaystyle \{p_{1}, p_{2}, \ldots, p_{n}\}$ poartă numele de *polii* polinomului $H$.

### 8.1 Conversia către răspunsul frecvență

Odată ce transformat $z$ a fost calculată putem defini *răspunsul frecvență* al unui filtru care este caracterizat de ecuațiile de tip LCCDE.

Pentru aceasta vom face schimbarea $z \rightarrow e^{j\omega}$. Așadar:

$$H(\omega)=\frac{\displaystyle \sum^{M}_{k=0}{b_{k}e^{-j\omega k}}}{\displaystyle \sum^{N}_{k=0}{a_{k}e^{-j\omega k}}}.$$

## 9. Filtre de tipul FIR (Finite Impulse Response)

Filtrele de tipul FIR sunt caracterizate de LCCDE: $$y[n]=\sum^{M-1}_{k=0}{b_{k}u[n-k]}, $$ unde $M$ poartă denumirea de *lungimea filtrului*, $M-1$ reprezintă *ordinul filtrului*, iar $b_{k} \in \mathbb{R}$ sunt coeficienții filtrului care descriu dependența de input-urile anterioare sau input-ul prezent. Folosinde-ne de ecuațiile anterioare deducem că:

$$H(z)=\sum^{M-1}_{k=0}{h[k]z^{-k}} \longrightarrow H(\omega)=\sum^{M-1}_{k=0}{h[k]e^{-j\omega k}}=\sum^{M-1}_{k=0}{b_{k}e^{-j\omega k}},$$ unde $z=e^{j\omega}.$

### 9.1 Filtrul Moving Average (MA)

Filtrul de tip MA folosește informația actuală și cea trecută pentru a produce un anumit output. Matematic putem rescrie: 

$$y[n]=\frac{1}{M}\sum^{M-1}_{k=0}{u[n-k]}$$ și făcând legătura cu ecuațiile de mai sus putem deduce că $\displaystyle b_{k}=\frac{1}{M}, (\forall) \text{ }k\in \{0,1,\ldots,M-1\}$.

Răspunsul frecvență pentru acest tip de filtru ce calculează după cum urmează:
$$H(\omega)=\frac{1}{M}\sum^{M-1}_{k=0}{e^{-j\omega k}}=\frac{1}{M}\sum^{M-1}_{k=0}{(e^{-j\omega})^{k}},$$ de unde urmează și concluzia că $$H(\omega)=\frac{1}{M}\cdot\frac{1-e^{-j\omega M}}{1-e^{-j\omega}}.$$


#### 9.1.1 Filtrul Moving Average. Exemplu practic.
Vom considera un filtru de lungime $M=3$ și coeficienții $\displaystyle b_{k}=\frac{1}{3}, (\forall)\text{ } k\in \{0,1,2\}.$ Așadar, ecuția care descrie comportamentul filtrului este: $$y[n]=\frac{1}{3}(u[n-2]+u[n-1]+u[n]).$$
În continuare aplicând transformata $z$ obținem: 

\begin{eqnarray} 
Y(z)&=&\frac{1}{3}(z^{-2}U(z)+z^{-1}U(z)+U(z))\\ 
&=& \frac{1}{3}(z^{-2}+z^{-1}+1)U(z).
\end{eqnarray} 

Această ultimă relație conduce la funcția de transfer:

$$H(z)=\frac{Y(z)}{U(z)}=\frac{1}{3}(1+z^{-1}+z^{-2}).$$

Următorul pas este să găsim *răspunsul frecvență* al filtrului. Acest lucru îl putem face folosind $z=e^{j\omega}$ în relația de mai sus.

\begin{eqnarray}
H(\omega) &=& \frac{1}{3}(e^{-j2\omega}+e^{-j\omega}+1)\\
&=& \frac{1}{3}e^{-j\omega}(e^{-j\omega}+e^{j\omega}+1)\\
&=&\frac{1}{3}e^{-j\omega}(2\frac{e^{-j\omega}+e^{j\omega}}{2}+1)\\
&=& \frac{1}{3}e^{-j\omega}(2\cos({\omega})+1).
\end{eqnarray}

În continuare avem că $\displaystyle \left|e^{-j\omega}\right|=1$ și că magnitudinea este $$H(\omega)=\frac{1}{3}\left|1+2\cos{(\omega)}\right|.$$ Se observă că pentru $\displaystyle \omega = 0 \Rightarrow H(\omega)=1$; $\displaystyle \omega = \arccos\left(-\frac{1}{2}\right)=\frac{2\pi}{3} \Rightarrow H(\omega)=0$; $\displaystyle \omega = \pi \Rightarrow H(\omega)=\frac{1}{3}$. Deducem din relațiile anterioare că faza este: 

$$\angle H(\omega) =  \begin{cases} 
      -\omega & \omega \in \left(0,\frac{2\pi}{3}\right) \\
      -\omega + \pi & \omega \in \left(\frac{2\pi}{3}, \pi\right)
   \end{cases}
$$

În continuare considerăm semnalul input ca fiind $$u[n]=\sin{\left(\frac{\pi}{10}n\right)}+\sin{\left(\frac{\pi}{4}n\right)}+\sin{\left(\frac{3\pi}{4}n\right)}$$
Cunoscând forma generală a magnitudinii putem calcula magnitudinea pentru fiecare componentă; astfel: 
* $$\left|H\left(e^{\frac{j\pi}{10}}\right)\right|\approx 0.967$$
* $$\left|H\left(e^{\frac{j\pi}{4}}\right)\right|\approx 0.805$$
* $$\left|H\left(e^{\frac{3j\pi}{4}}\right)\right|\approx 0.138$$

Tinând cont de magnitudinea de pe fiecare componentă, precum și de faza shiftată, forma finală a semnalului filtrat este: 

$$y[n]=0.967\sin{\left(\frac{n\pi}{10}-\frac{\pi}{10}\right)}+0.805\sin{\left(\frac{n\pi}{4}-\frac{\pi}{4}\right)}+0.138\sin{\left(\frac{3n\pi}{4}-\frac{3\pi}{4}\right)}.$$

Din relația de mai sus reținem că componentele cu frecvență mai mică sunt reținute/folosite mai mult în comparație cu componenta de frecvență mai mare.

```python
import numpy as np
import matplotlib.pyplot as plt


N = 100  #samples 

#input signal
n = np.arange(N)
u_n = np.sin((np.pi / 10) * n) + np.sin((np.pi / 4) * n) + np.sin((3 * np.pi / 4) * n)

#frequnecies
frequencies = [np.pi / 10, np.pi / 4, 3 * np.pi / 4]

def H_omega(omega):
    """Evaluate the frequency response H(omega) for a given omega."""
    return (1 / 3) * np.exp(-1j * omega) * (2 * np.cos(omega) + 1)


magnitudes = [np.abs(H_omega(omega)) for omega in frequencies]
phases = [np.angle(H_omega(omega)) for omega in frequencies]

#filtered output y[n] 
y_n = (magnitudes[0] * np.sin(frequencies[0] * n + phases[0]) +
       magnitudes[1] * np.sin(frequencies[1] * n + phases[1]) +
       magnitudes[2] * np.sin(frequencies[2] * n + phases[2]))


omega = np.linspace(0, np.pi, 250) 
H_omega_vals = H_omega(omega)
magnitude_H = np.abs(H_omega_vals)
phase_H = np.angle(H_omega_vals)


plt.figure(figsize=(12, 10))

plt.subplot(3, 1, 1)
plt.plot(n, u_n, label=r'Semnalul input $u[n]$', color='b')
plt.plot(n, y_n, label=r'Semnalul filtrat $y[n]$', color='r')
plt.title('Semnalul input. Semnalul filtrat')
plt.xlabel('n')
plt.ylabel('Amplitudine')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(omega, magnitude_H, color='g')
plt.title('Răspunsul magnitudine $|H(\omega)|$')
plt.xlabel(r'$\omega$ (radiani)')
plt.ylabel('Magnitudine')
plt.grid(True)


plt.subplot(3, 1, 3)
plt.plot(omega, phase_H, color='purple')
plt.title('Răspunsul fază $H(\omega)$')
plt.xlabel(r'$\omega$ (radiani)')
plt.ylabel('Faza (radiani)')
plt.grid(True)


plt.tight_layout()
plt.show()
```
